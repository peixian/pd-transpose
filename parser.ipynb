```python
import pandas as pd
```

```python
dfs = {}
for s in ['children-injured', 'children-killed', 'injured', 'killed']:
    df = pd.read_csv(f"{s}.csv")
    df = df.drop(['Totals'], axis=1, errors='ignore') # drop totals if it exists
    df = df.transpose()
    df.columns = df.iloc[0] 
    df = df[1:]
    dfs[s] = df
```

```python
dfs['children-injured'].head()
```

```python
# for every column, get the non-nan, and create a new DF that has them
def count_cities(df):
    civcas_totals = pd.DataFrame()
    for col in df.columns:
        civcas_dates = df[df[col].notnull()][col].reset_index()
        if len(civcas_dates) > 0:
            #print(civcas_dates)
            civcas_dates['date'] = pd.to_datetime(civcas_dates['index'], format='%d-%b')
            civcas_dates['date'] = civcas_dates['date'].apply(lambda x: x.replace(year = 2022))
            civcas_dates['location'] = col
            civcas_dates['count'] = civcas_dates[col]
            #print(civcas_dates[[col, 'date', 'location']])
            civcas_totals = civcas_totals.append(civcas_dates[['count', 'date', 'location']])
    civcas_totals = civcas_totals[['date', 'location', 'count']]
    return civcas_totals
```

```python
count_cities(dfs['children-injured'])
for k, v in dfs.items():
    count_cities(v).to_csv(f"{k}-parsed.csv", index=False)
```
